#### 2. Shoot randomly until all ships are sunken

```python
import numpy as np

env = BattleshipsBoard()
env.randomly_place_all_ships()

rng = np.random.default_rng()
finished = False
steps = 0

while not finished:
    # pick a random untried cell
    candidates = np.argwhere(env.attacks == 0)   # shape (k, 2) with [y, x]
    y, x = candidates[rng.integers(len(candidates))]

    result = env.receive_attack(y, x)
    finished = result.finished
    steps += 1

print("Finished in", steps, "shots")
env.visualize()
```
